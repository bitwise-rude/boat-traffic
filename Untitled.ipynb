```python
## 1D
train = df.iloc[:-12]
test  = df.iloc[-12:]

y_train = train['Boat_Imputed']
y_test  = test['Boat_Imputed']
```

```python
# HOLT winners (Assesement 2)
from statsmodels.tsa.holtwinters import ExponentialSmoothing

hw_model = ExponentialSmoothing(
    y_train,
    trend='add',
    seasonal='add',
    seasonal_periods=12
).fit()

hw_forecast = hw_model.forecast(12)
```

```python
plt.figure(figsize=(10,4))
plt.plot(y_train, label="Train")
plt.plot(y_test, label="Test")
plt.plot(hw_forecast, label="HW Forecast")
plt.legend()
plt.title("Holt-Winters Forecast")
plt.show()
```

```python
# MASE calculation (2b)
def mase(y_true, y_pred, y_train, m=12):
    naive = y_train.shift(m)
    d = (y_train[m:] - naive[m:]).abs().mean()
    return (y_true - y_pred).abs().mean() / d



hw_mase = mase(
    y_true=y_test,
    y_pred=hw_forecast,
    y_train=y_train
)

hw_mase
```

```python
# Prophet (Assesement 3)
prophet_df = df[['Boat_Imputed']].reset_index()
prophet_df.columns = ['ds', 'y']

train_p = prophet_df.iloc[:-12]
test_p  = prophet_df.iloc[-12:]
```

```python
# Additivie v Multi
from prophet import Prophet
from sklearn.metrics import mean_squared_error

def prophet_rmse(seasonality):
    m = Prophet(seasonality_mode=seasonality)
    m.fit(train_p)
    future = m.make_future_dataframe(periods=12, freq='MS')
    forecast = m.predict(future)
    yhat = forecast['yhat'].iloc[-12:].values
    return np.sqrt(mean_squared_error(test_p['y'], yhat)), m, forecast

rmse_add, model_add, fc_add = prophet_rmse('additive')
rmse_mul, model_mul, fc_mul = prophet_rmse('multiplicative')

rmse_add, rmse_mul
```

```python
# Trend Plotting
chosen_model = model_add if rmse_add < rmse_mul else model_mul
chosen_model.plot(fc_add if chosen_model==model_add else fc_mul)
plt.show()

chosen_model.plot_components(fc_add if chosen_model==model_add else fc_mul)
plt.show()
```

```python
# Assement 3d ACF/PACF of Remainder
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

residuals = train_p['y'] - chosen_model.predict(train_p)['yhat']

plot_acf(residuals.dropna(), lags=24)
plt.show()

plot_pacf(residuals.dropna(), lags=24)
plt.show()
```
